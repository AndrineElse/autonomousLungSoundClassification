```python
sorted_2class = {k: v for k, v in sorted(performance_time_2class.items(), key=lambda item: item[1])}
color_list = [color_dict[k] for k,v in sorted_2class.items()]
n_tsc = len(sorted_2class)


f, ax = plt.subplots(1,1)
f.set_size_inches(6, 2)
ax.bar(*zip(*sorted_2class.items()))
ax.set_yscale('log')

for label in ax.get_yticklabels() :
    #print(label)
    label.set_fontproperties(font_small)
    
for label in ax.get_xticklabels() :
    label.set_fontproperties(font_small)
    label.set_rotation(50)
    
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
    
ax.set_ylabel('Time [s]', fontproperties = font)

count = 0
for child in ax.get_children():
    if (type(child) == matplotlib.patches.Rectangle) and (count < (n_tsc)):
        child.set_color(color_list[count%n_tsc])
        #child.set_alpha(0.8)
        count = count +1

        
#ax.get_legend().remove()

f.savefig(figure_path + f'compuation_time_plot_2class.pdf', bbox_inches='tight')
```
